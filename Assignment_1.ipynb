```python
Q3. Describe three different data types.
Ans. 
    1. Numeric: these data types are used to store numeric values:
    i)integer: store integer values
    ii)float: stores floating point numbers 
    iii)complex: used to store complex numbers 
    2. Sequence data types: these data types are used to store ordered collection of similar  or different data.
    i)string: ordered collection of characters.
    ii)list: array of similar or different type of data. 
    iii)tuples: they are same like array, only difference is, they are immutable.
    3. Boolean: it is a data type that contain one of the two built in values; i.e. true or false
```
